# Passo a passo do Problema

- Passo1: buscar dados das ações automaticamente
- Passo2: realizar as análises solicitadas pelo meu gestor
- Passo3: enviar automaticamente o email com resultado da análise

# Passo 1

```python
import yfinance as yf

ticker =  input("Digite o código da açáo: ")
dados = yf.Ticker(ticker).history("6mo")
fechamento = dados.Close
fechamento
```

```python
fechamento.plot()
```

# Passo 2 - realizar as análises solicitadas pelo meu gestor

- cotação máxima
- cotação minima
- cotação atual

```python
maxima = round(fechamento.max(), 2)
minima = round(fechamento.min(), 2)
atual = round(fechamento[-1] ,2 )

print(maxima)
print(minima)
print(atual)
```

# Passo 3 -  enviar automaticamente o email com resultado da análise

- abrir uma nova aba (ctrl + t)
- digitar email e dar um enter
- clicar no botão escrever 
- digitar o destinatario (tab)
- preencher o corpo do e-mail
- clicar no botão enviar

```python
import pyautogui
import pyperclip
```

```python
# dar pausa 3 segundos entre os passos
pyautogui.PAUSE = 3

# abrir uma nova aba (ctrl + t)
pyautogui.hotkey("ctrl", "t")

#digitar email e dar um enter
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

#clicar botão escrever
pyautogui.click(x=-1367, y=-123)

#digitar o destinatario (tab)
pyperclip.copy("jm.ubaldo@hotmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

#digitar o assunto (tab)
pyperclip.copy("Analise diárias")
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("tab")

#preencher o corpo do e-mail
mensagem = f"""
Prezado gestor,

Seguem, conforme solicitado, as análises dos últimos 6 meses da ação {ticker}:

Cotação máxima: R$ {maxima}
Cotação mínima: R$ {minima}
Cotação atual: R$ {atual}

Qualquer dúvuda, fico à disposição!

Atte.
Joel Ubaldo

"""

pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl","v")

#enviar e-mail
pyautogui.click(x=-619, y=545)
print("e-mail enviado com sucesso!")



```
